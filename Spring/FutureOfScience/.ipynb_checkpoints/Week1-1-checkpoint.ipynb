## 데이터 생성 및 시각화
numpy 라이브러리를 사용해 임의의 데이터를 생성, matplotlib로 이를 시각화

```python
import numpy as np
from matplotlib import pyplot as plt

# 임의의 데이터 생성
ys = 200 + np.random.randn(100) # 200+-a의 난수 100개 배열 생성 
x = [x for x in range(len(ys))] # y 데이터 길이로 0~99 값 배열 생성

plt.plot(x, ys, '-') # x, y로 이루어진 꺾은선 그래프 생성
# 그래프 아래를 반투명 초록색으로 채움 
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Sample Visualization") # 차트 제목 
plt.show()
```

## Array

```python
np.random.seed(0) # seed for reproducibility 

# 10 미만의 수로 이루어진 array 생성
arr1 = np.random.randint(10, size=6) # 개수 6인 one dimensional array
arr2 = np.random.randint(10, size=(3, 4)) # 3행 4열인 two dimensional array
arr3 = np.random.randint(10, size=(3, 4, 5)) # 3 by 4 by 5 three dimensional array
```

```python
print(arr1)
```

```python
print(arr2)
```

```python
print(arr3)
```

```python
# Array의 다양한 정보 출력하기
print("type of arr3: ", type(arr3))
print("ndim of arr3: ", arr3.ndim)
print("shape of arr3: ", arr3.shape)
print("size of arr3: ", arr3.size)
print("dtype of arr3: ", arr3.dtype) # data type
```

```python
# Create Array via list
A = np.array([[1,2,3], [3,4,5]]) 
print(A)

A = np.array([[1,2,3], [3,4.4, 5]]) # array of floats 
print(A)
```

```python
# Array of Zeros and Ones
zeros_array = np.zeros((2,3)) # 2,3 shape
print(zeros_array)

ones_array = np.ones((1,5), dtype=np.int32) # 1,5 shape = vector
print(ones_array)
```

```python
# Using arange() and reshape()
A = np.arange(4) # 0 ~ 3
print('A=', A)

B = np.arange(12)
C = B.reshape(2, 6) # B를 2, 6으로 변환 
print('B =', B)
print(C)
```

```python
# Transpose of a Matrix
print(C.T) 
print(C.transpose())
```

```python
# Access matrix elements, rows & columns
A = np.array([2, 4, 6, 8, 10])
print("A[0] =", A[0]) # First Element
print("A[-1] =", A[-1]) # Last Element

A = np.array([[1, 4, 5, 12], [-5, 8, 9, 0], [-6, 7, 11, 19]]) # 2D array
print("A[0][0] =", A[0][0]) # 1행 1열 원소
print("A[1][2] =", A[1][2]) # 2행 3열 원소

# :은 해당 dimension의 정보를 모두 사용하겠다는 뜻
print("A[:, 0] =", A[:, 0]) # 1열 전체
print("A[:, 0] =", A[:, 3]) # 4열 전체
print("A[:, 0] =", A[0, :]) # 1행 전체
```

```python
# Slicing of a Matrix
letters = np.array([1, 3, 5, 7, 9, 5])

print(letters[2:5]) # index 2 ~ 4 원소
print(letters[:])
print(letters[::-1]) # reversing a list
```

## Image Data

```python
from scipy import misc

f = misc.face() # face 영상 가져와 저장
```

```python
print(f.shape) # 768행 1024열 3채널
print(f.max()) # 최대 값 255 -> 8bit
print(f.dtype)
# print(f)
```

```python
plt.imshow(f) # 이미지 시각화 메서드
plt.axis("off") # remove the axis and the ticks 
plt.show()
```

```python
plt.imshow(f[:, :, 0], cmap=plt.cm.gray) # Red channel
plt.axis("off") # remove the axis and the ticks 
plt.show()
```

```python
plt.imshow(f[:, :, 1], cmap=plt.cm.gray) # Green channel
plt.axis("off") # remove the axis and the ticks 
plt.show()
```

```python
plt.imshow(f[:, :, 2], cmap=plt.cm.gray) # Blue channel
plt.axis("off") # remove the axis and the ticks 
plt.show()
```

```python
data = {'apples' : 10, 'oranges': 15, 'lemons': 5, 'limes': 20}
names = list(data.keys())
values = list(data.values())

fig, axs = plt.subplots(1, 3, figsize=(9, 3), sharey=True) # 3개의 그래프 한 줄에 출력
axs[0].bar(names, values) 
axs[1].scatter(names, values)
axs[2].plot(names, values)
fig.suptitle('Categorical Plotting')
plt.show()
```
